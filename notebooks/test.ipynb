```python
import tkinter as tk
import main

class ConfirmationWindow(tk.Toplevel):
    def __init__(self, languages, difficulties):
        super().__init__()

        self.title("选项确认")
        self.languages = languages
        self.difficulties = difficulties

        tk.Label(self, text="您选择的所学语言是：").pack(anchor=tk.W)
        for lang in languages:
            tk.Label(self, text=lang).pack(anchor=tk.W)

        tk.Label(self, text="您所选的学习难度是：").pack(anchor=tk.W)
        for diff in difficulties:
            tk.Label(self, text=diff).pack(anchor=tk.W)

        confirm_button = tk.Button(self, text="Confirm", command=self.open_main_window)
        confirm_button.pack()

    def open_main_window(self):
        self.destroy()
        MainWindow()

class LanguageSelection(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("请选择你想学习的语言")

        self.languages = {
            "英语": tk.IntVar(),
            "法语": tk.IntVar(),
            "德语": tk.IntVar(),
            "西班牙语": tk.IntVar(),
            "日语": tk.IntVar(),
            "韩语": tk.IntVar()
        }

        tk.Label(self, text="选择语言：").pack(anchor=tk.W)
        for lang, var in self.languages.items():
            tk.Checkbutton(self, text=lang, variable=var).pack(anchor=tk.W)

        self.next_button = tk.Button(self, text="下一步", command=self.show_difficulty_selection)
        self.next_button.pack()

    def show_difficulty_selection(self):
        selected_languages = [lang for lang, var in self.languages.items() if var.get() == 1]
        if not selected_languages:
            tk.messagebox.showerror("错误！", "请至少选择一种语言！")
            return
        if len(selected_languages) >= 2:
            tk.messagebox.showerror("错误！", "只能选择一种语言！")
            return

        self.destroy()

        DifficultySelection(selected_languages)

class DifficultySelection(tk.Toplevel):
    def __init__(self, languages):
        super().__init__()

        self.title("难度选择")

        self.difficulties = {
            "日常会话": tk.IntVar(),
            "商务交流": tk.IntVar(),
            # Add more options as needed
        }

        tk.Label(self, text="难度选择：").pack(anchor=tk.W)
        for diff, var in self.difficulties.items():
            tk.Checkbutton(self, text=diff, variable=var).pack(anchor=tk.W)

        self.confirm_button = tk.Button(self, text="确认", command=lambda: self.open_confirmation_window(languages))
        self.confirm_button.pack()

    def open_confirmation_window(self, languages):
        selected_difficulties = [diff for diff, var in self.difficulties.items() if var.get() == 1]
        if not selected_difficulties:
            tk.messagebox.showerror("错误！", "请至少选择一种难度！")
            return
        if len(selected_difficulties) >= 2:
            tk.messagebox.showerror("错误！", "只能选择一种难度！")
            return

        self.destroy()

        ConfirmationWindow(languages, selected_difficulties)

class MainWindow(tk.Tk):
    def __init__(self):
        super().__init__()

        # self.title("主程序窗口")

        main.runMySeg()
        main.printInfo()
        # Your main window content goes here

if __name__ == "__main__":
    app = LanguageSelection()
    app.mainloop()
```
